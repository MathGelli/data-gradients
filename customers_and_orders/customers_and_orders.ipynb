**Question 1:** How many customers are in the file?
(as help, we have added some comments and starter code to help you structure the solution)

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

with open(datafile, 'r') as f1:
    csvreader = csv.reader(f1, delimiter=',')

    next(csvreader)  # pula o cabeçalho

    count = 0
    for row in csvreader:
        count += 1

print("Number of customers:", count)
```

**Question 2:** In how many different states do the customers live in?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

with open(datafile, 'r') as f1:
    csvreader = csv.DictReader(f1)

    states = set()
    for row in csvreader:
        state = row['State'].strip().upper()
        states.add(state)

print("Different states:", len(states))
print(states)
```

**Question 3** What is the state with most customers?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

state_count = {}

with open(datafile, 'r') as f1:
    csvreader = csv.DictReader(f1)

    for row in csvreader:
        state = row['State'].strip().upper()
        state_count[state] = state_count.get(state, 0) + 1

most_state = max(state_count, key=state_count.get)

print("State with most customers:", most_state)
print(state_count)
```

**Question 4** What is the state with the least customers?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

state_count = {}

with open(datafile, 'r') as f1:
    csvreader = csv.DictReader(f1)

    for row in csvreader:
        state = row['State'].strip().upper()
        state_count[state] = state_count.get(state, 0) + 1

min_count = min(state_count.values())
least_states = [state for state, count in state_count.items() if count == min_count]

print("Least number of customers:", min_count)
print("States with least customers:", least_states)
print(state_count)
```

```python
from os.path import exists
import csv

datafile = "./sample_data/orders.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

order_counts = {}

with open(datafile, 'r') as f1:
    csvreader = csv.DictReader(f1)

    for row in csvreader:
        order_id = row['OrderID']
        order_counts[order_id] = order_counts.get(order_id, 0) + 1

average_items = sum(order_counts.values()) / len(order_counts)

print("Average items per order:", round(average_items, 2))
```

```python
from os.path import exists
import csv

datafile = "./sample_data/orders.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

october_orders = set()

with open(datafile, 'r') as f1:
    csvreader = csv.DictReader(f1)

    for row in csvreader:
        order_date = row['Date']
        order_id = row['OrderID']

        if order_date.startswith('2021-10'):
            october_orders.add(order_id)

print("Orders placed in October 2021:", len(october_orders))
```

```python
from os.path import exists
import csv

orders_file = "./sample_data/orders.csv"
customers_file = "./sample_data/customers.csv"

if not exists(orders_file) or not exists(customers_file):
    raise SystemExit("You should run the first code cell and download the dataset files!")

# 1) carregar nomes dos clientes
customers = {}
with open(customers_file, "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        customer_id = row["CustomerID"].strip()
        full_name = f"{row['FirstName'].strip()} {row['LastName'].strip()}"
        customers[customer_id] = full_name

# 2) somar gasto por cliente em 2021, contando cada pedido uma vez
spent_by_customer = {}
seen_orders = set()

with open(orders_file, "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        order_id = row["OrderID"].strip()
        customer_id = row["CustomerID"].strip()
        order_date = row["Date"].strip()
        order_total = float(row["OrderTotal"])

        # apenas pedidos de 2021
        if not order_date.startswith("2021"):
            continue

        # evita contar o mesmo pedido mais de uma vez
        if order_id in seen_orders:
            continue

        seen_orders.add(order_id)
        spent_by_customer[customer_id] = spent_by_customer.get(customer_id, 0) + order_total

# 3) achar o cliente que mais gastou
top_customer_id = max(spent_by_customer, key=spent_by_customer.get)
top_amount = spent_by_customer[top_customer_id]
top_name = customers.get(top_customer_id, "Unknown customer")

print("Customer who spent the most in 2021:")
print(top_name, top_customer_id, top_amount)
```

```python
from os.path import exists
import csv

orders_file = "./sample_data/orders.csv"

if not exists(orders_file):
    raise SystemExit("You should run the first code cell and download the dataset files!")

month_names = {
    "01": "January",
    "02": "February",
    "03": "March",
    "04": "April",
    "05": "May",
    "06": "June",
    "07": "July",
    "08": "August",
    "09": "September",
    "10": "October",
    "11": "November",
    "12": "December"
}

sales_by_month = {}
seen_orders = set()

with open(orders_file, "r") as f:
    reader = csv.DictReader(f)

    for row in reader:
        order_id = row["OrderID"].strip()
        order_date = row["Date"].strip()
        order_total = float(row["OrderTotal"])

        # conta cada pedido só uma vez
        if order_id in seen_orders:
            continue

        seen_orders.add(order_id)

        month = order_date[5:7]   # pega o mês da data YYYY-MM-DD
        sales_by_month[month] = sales_by_month.get(month, 0) + order_total

best_month_number = max(sales_by_month, key=sales_by_month.get)
best_month_name = month_names[best_month_number]

print("Best month for sales:", best_month_name)
print("Sales by month:", sales_by_month)
```
